---

## Evolucao em relacao a v1

| Aspecto | v1 | v2 |
|---|---|---|
| Fonte de dados | CSV local sem origem definida | SIGEO LOA SP (dado publico rastreavel) |
| Limpeza | dropna simples | validacao de schema, log de qualidade, copia imutavel |
| Encoding | nenhum | LabelEncoder para variavel categorica orgao |
| Modelo | LinearRegression basica | LinearRegression com persistencia joblib |
| Forecast | ausente | previsao por orgao para proximo exercicio |
| Rastreabilidade | ausente | celula de metadata com fonte, data e versao |
| Governanca | ausente | dados brutos preservados, .gitignore, requirements.txt |
